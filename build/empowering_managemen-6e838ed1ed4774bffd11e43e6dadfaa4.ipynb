# エンパワメント型マネジメント

メンバーがパフォーマンスを発揮させられるように場を整えることに徹するマネジメントスタイル

Googleが提唱するスタイル

### 効果的なチーム

[Google re:Work - ガイド: 「効果的なチームとは何か」を知る](https://rework.withgoogle.com/intl/jp/guides/understanding-team-effectiveness)

1. **サイコロジカル・セーフティー（心理的安全性）**：チームメンバーがリスクを取ることを安全だと感じ、お互いに対して弱い部分もさらけ出すことができる
2. **相互信頼**：チームメンバーが他のメンバーが仕事を高いクオリティで時間内に仕上げてくれると感じている
3. **構造と明確さ**：チームの役割、計画、目標が明確になっている
4. **仕事の意味**：チームメンバーは仕事が自分にとって意味があると感じている
5. **インパクト**：チームメンバーは自分の仕事について、意義があり、良い変化を生むものだと思っている


## エンパワメント型マネージャー

- 明確なビジョンを示し、チームと野心的な目標を共有する
- **率先垂範** ：自ら達成にコミットする。メンバーを徹底してサポートする
- 可能な限りメンバーに権限委譲し、メンバーが自律的に動ける環境を整える
- マネージャーが変化・成長する：現状維持ではなく、自分がどうあるべきか、どう成長していくかを常に考える
- **コーチングする** ：解決策を与えるのではなく、メンバーが自身で答えにたどり着けるように促す

### 目標設定

1. マネージャーがメンバーに対し、 **ビジネス戦略、チームの目標・ゴール、メンバーへの「期待」** をクリアに伝える
2. メンバーは、 **具体的にどういう行動をとり、どのような成果を出すのか** を自分で考える。
3. マネージャーは「なぜそのアクションを取れば達成できるといえるのか」を議論し、アクションをすり合わせる
