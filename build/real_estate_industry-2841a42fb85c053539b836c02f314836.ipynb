# 不動産業界

## 不動産業界のビジネスモデル

大別して

1. 開発・企画・建設（デベロッパー）
2. 流通
3. 賃貸管理
4. 建物管理

に分かれる

### 1. 開発・企画・建設（デベロッパー）

土地の仕入れ、マンション・商業施設・宅地などの企画・開発・建設・販売を行う。

大資本が必要なため大手企業が多く、街づくりに深く関わる。


### 2. 流通

#### 仲介業

売買取引や賃貸取引の仲介を行う。

例：三井のリハウス、ノムコム（ https://www.nomu.com/ ）


#### 買取再販

仲介業者から物件を買いとり、リノベーション等の独自の付加価値を乗せて販売する

例：カチタス、GA technologies


### 3. 賃貸管理

不動産のオーナー（大家さん）から賃貸周りの管理を委託され、入居者探し、家賃回収、入居者からの問い合わせ対応などを請け負う。

例：東急リバブル

### 4. 建物管理

集合住宅の共用部の管理を請け負う。日々の清掃を行う管理人の手配や、定期的なメンテナンス、大規模修繕など。

例：TFDコミュニティ、スカイサービス
