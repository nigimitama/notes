# LangExtractパッケージ

LangExtract は Googleによる情報抽出のためのPythonパッケージ

:::{card} 公式GitHubリポジトリ

https://github.com/google/langextract

:::


## Why LangExtract?

VLM（LLM）による情報抽出は手軽にパフォーマンスを出せるようになってきた。

ただし、実務で使うためには、ハルシネーションのリスクを軽減するために「文書のどこから取得したのか（source grounding）」を調べて検証できるようにしたりなど、細かい部分を整える必要がある。

そこを簡単にできるようにするのがLangExtractパッケージ
