# ベクトル空間

ベクトル空間：実数倍と和が定義されており、任意の元の線形結合がふたたび元となる集合

:::{admonition} 定義（ベクトル空間）

任意の元$x, y \in L$と任意の実数$a, b$について実数倍$ax, by$とそれらの和$ax + by$が定義されており、
かならず$ax + by \in L$が成り立つような集合$L$を **ベクトル空間（vector space）** という。また$L$の元を **ベクトル（vector）** という
:::

イメージ的には原点と矢印と、矢印同士の演算（和や積など）だけが決まっている。座標は固定ではない

整数集合$\mathbb{Z}$は実数を掛けて線形結合した結果が整数になるとは限らないため、ベクトル空間ではない。
実数集合$\mathbb{R}$はベクトル空間である。


:::{dropbox} （参考）和と定数倍の定義まで含めた細かい公理

空でない集合 $\mathcal{L}$ が次の性質を持つとき、 これを **線形空間** （または **ベクトル空間** ） と呼ぶ。

(1) 任意の $\boldsymbol{u}, \boldsymbol{v} \in \mathcal{L}$ に対して **和** と呼ばれる元 $\boldsymbol{u}+\boldsymbol{v} \in \mathcal{L}$ がただ一つ存在し、次の性質を持つ。

1. 任意の $\boldsymbol{u}, \boldsymbol{v} \in \mathcal{L}$ に対して $\boldsymbol{u}+\boldsymbol{v}=\boldsymbol{v}+\boldsymbol{u}$ (和の交換法則)
2. 任意の $\boldsymbol{u}, \boldsymbol{v}, \boldsymbol{w} \in \mathcal{L}$ に対して $\boldsymbol{u}+(\boldsymbol{v}+\boldsymbol{w})=(\boldsymbol{u}+\boldsymbol{v})+\boldsymbol{w}$ (和の結合法則)
3. 任意の $u, v \in \mathcal{L}$ に対して $u+x=v$ となる元 $x \in \mathcal{L}$ が存在する（その元を $\boldsymbol{v}-\boldsymbol{u}$ と書く)（和の逆元）

(2) 任意の元 $\boldsymbol{u} \in \mathcal{L}$ と任意の実数 $\lambda$ に対して **$\lambda$ 倍** と呼ばれる元 $\lambda \boldsymbol{u} \in \mathcal{L}$ がただ一つ存在し、次の性質を持つ。

1. 任意の $\boldsymbol{u}, \boldsymbol{v} \in \mathcal{L}$ と任意の実数 $\lambda$ に対して $\lambda(\boldsymbol{u}+\boldsymbol{v})=\lambda \boldsymbol{u}+\lambda \boldsymbol{v}$
2. 任意の $\boldsymbol{u} \in \mathcal{L}$ と任意の実数 $\lambda, \mu$ に対して $(\lambda+\mu) \boldsymbol{u}=\lambda \boldsymbol{u}+\mu \boldsymbol{u}$
3. 任意の $\boldsymbol{u} \in \mathcal{L}$ と任意の実数 $\lambda, \mu$ に対して $(\lambda \mu) \boldsymbol{u}=\lambda(\mu u)$
4. 任意の $u \in \mathcal{L}$ に対して $1 u=u$

:::

## 部分空間

:::{admonition} （定義）部分ベクトル空間

ベクトル空間 $V$ の空でない部分集合 $W$ が $V$ における和とスカラー倍の演算によってベクトル空間になるとき、すなわち

1. $W \neq \phi \quad$ ( $\phi$ は空集合)
    - 同値の別表現として$\boldsymbol{0} \in W$（$W$は零ベクトルを含む）が1番目の条件にくる定義もある
2. $\boldsymbol{a}, \boldsymbol{b} \in W \Rightarrow \boldsymbol{a}+\boldsymbol{b} \in W$
3. $\boldsymbol{a} \in W, \quad \lambda \in \boldsymbol{R} \Rightarrow \lambda \boldsymbol{a} \in W$

を満たすとき、 $W$ を $V$ の **部分空間** 、または **部分べクトル空間** という。
:::

:::{card} 部分空間の例

- ベクトル空間$V$自身や$V$の零元だけからなる集合$\{0\}$
- $K$上のベクトル空間$V$の任意の元$v$に対して、集合$\{av | a \in K\}$
- $\mathbb{R}^n$に対し、原点を含む直線、平面、超平面
    - 原点を含まない直線、平面、超平面は **アフィン部分空間**
:::

:::{card} 部分空間の例：Im, Ker

Kerについては：

$V, W$をベクトル空間とし、$f:V\to W$を線形写像とする。

$x_1, x_2\in \operatorname{Ker} f$とする。Kerの定義$\operatorname{Ker}f :=\{ x\in V \mid f(x) = 0 \}$から

$$
f(x_1) = f(x_2) = 0
$$

線形写像の線形性から、$c_1, c_2 \in \mathbb{R}$について

$$
f(c_1 x_1 + c_2 x_2)
= c_1 f(x_1) + c_2 f(x_2)
= 0
$$

なので$c_1 x_1 + c_2 x_2 \in \operatorname{Ker} f$ であり、和と定数倍に閉じていて空集合でないため$\operatorname{Ker} f$は$V$の部分空間である。

（詳細：[線形写像の像(Im),核(Ker)の定義とそれが部分空間になる証明 | 数学の景色](https://mathlandscape.com/im-ker/)）
:::

## スパン（ベクトルが張る空間）

ベクトル空間$V$から選ばれたいくつかのベクトル$\boldsymbol{x}_1, \dots, \boldsymbol{x}_n$を用いて作られる部分集合

:::{admonition} （定義）スパン

ベクトル$\boldsymbol{x}_1, \dots, \boldsymbol{x}_n$（$\in V$）の線形結合$a_1 \boldsymbol{x}_1 + \cdots + a_n \boldsymbol{x}_n$の集合

$$
W = \{ a_1 \boldsymbol{x}_1 + \cdots + a_n \boldsymbol{x}_n  | a_1, \dots, a_n \in \mathbb{R} \}
$$

は部分空間であり、これを$\boldsymbol{x}_1, \dots, \boldsymbol{x}_n$の**スパン（span）**あるいは$\boldsymbol{x}_1, \dots, \boldsymbol{x}_n$**が張る空間**といい

$$
\text{Span}\{ \boldsymbol{x}_1, \dots, \boldsymbol{x}_n \}
$$

と書く
:::


#### （例）線形回帰モデルとの関わり

$\boldsymbol{x}_1$を「教育を受けた年数」の観測値のベクトル、$\boldsymbol{x}_2$を「その職種の経験年数」の観測値のベクトル、$\boldsymbol{y}$を「賃金」の観測値のベクトルとしたとき、$\text{Span}\{ \boldsymbol{x}_1, \boldsymbol{x}_2 \}$に$\boldsymbol{y}$が属するならば

$$
賃金 = a_1 \times 教育を受けた年数 + a_2 \times その職種の経験年数
$$

という関係が成り立つことになる




## 線形独立（1次独立）

:::{admonition} （定義）線形独立

$\boldsymbol{x}_1, \dots, \boldsymbol{x}_n \in V$について、係数$c_1,\dots,c_n$が

$$
c_1 = \cdots = c_n = 0
$$

であるとき、またそのときにのみ

$$
c_1 \boldsymbol{x}_1 + \cdots + c_n \boldsymbol{x}_n = \boldsymbol{0}
$$

が成り立つとき、$\boldsymbol{x}_1, \dots, \boldsymbol{x}_n$は **線形独立（linearly independent）** あるいは **1次独立** であるという。
:::

例えば、

$$
\begin{align}
\boldsymbol{x}_1 &= (1, 0, 1)^T\\
\boldsymbol{x}_2 &= (-1, 2, 3)^T
\end{align}
$$

とする。


$$
c_1 \boldsymbol{x}_1 + c_2 \boldsymbol{x}_2 = \boldsymbol{0}
$$

とすると、第2成分が0になるためには$c_2 = 0$である必要がある。

そして$c_1 \boldsymbol{x}_1 = \boldsymbol{0}$となるためには、$c_1 = 0$である必要がある。

よって$c_1 = c_2 = 0$であるため、$\boldsymbol{x}_1, \boldsymbol{x}_2$は1️次独立である。

:::{admonition} （定理）線形独立の同値な別表現

ベクトルの組$\boldsymbol{a}_1, \cdots, \boldsymbol{a}_n$について、次の(1)と(2)は同値である

(1) $\boldsymbol{a}_1, \cdots, \boldsymbol{a}_n$ は線形独立である

(2) $\boldsymbol{a}_1, \cdots, \boldsymbol{a}_n$ の線形結合（1次結合）で表わされる元の表わし方は一意的である。すなわち、

$$
c_1 \boldsymbol{a}_1+\cdots+c_n \boldsymbol{a}_n=c_1^{\prime} \boldsymbol{a}_1+\cdots+c_n^{\prime} \boldsymbol{a}_n
$$

ならば、 $c_1=c_1^{\prime}, \cdots, c_n=c_n^{\prime}$ である。
:::

:::{dropdown} 証明

(1) $\Rightarrow$ (2) について：

（2）の等式を移項する。

$$
\left(c_1-c_1{ }^{\prime}\right) \boldsymbol{a}_1+\cdots+\left(c_n-c_n{ }^{\prime}\right) \boldsymbol{a}_n=\mathbf{0}
$$

（1）の仮定より

$$
c_1-c_1^{\prime}=0, \cdots, c_n-c_n^{\prime}=0 \text { つまり } c_1=c_1^{\prime}, \cdots, c_n=c_n{ }^{\prime}
$$

であるから，表わし方は一意的である。

(2) $\Rightarrow$ (1)について：

1 次関係 $c_1 \boldsymbol{a}_1+\cdots+c_n \boldsymbol{a}_n=\mathbf{0}$ をみたす $c_i$ として, $c_i=0 \quad (i=1, \cdots, n)$ があるが，仮定よりこの他にはこの式をみたすものはない。

つまり $c_i=0 \quad(i=1, \cdots, n)$ が成り立ち, これは $a_1, \cdots, a_n$ が 1 次独立であることを意味する。

:::

::::{admonition} 例題

:::{card}
ベクトル

$$
\boldsymbol{a}_1=\left(\begin{array}{l}
2 \\
1
\end{array}\right), \boldsymbol{a}_2=\left(\begin{array}{l}
1 \\
2 \\
\end{array}\right)
$$

が1次独立かどうか判定せよ。
:::

行列のランク（行列$A$の列ベクトルの中から選び得る1次独立なベクトルの最大個数）を調べる方法がある。行列式がゼロでない（$|A|\neq 0$）なら$\operatorname{rank}(A)=n, A\in\mathbb{R}^{n\times n}$であるため、$A$が正方行列なら行列式がゼロかどうかを判定すればよい。

$$
|A|
=
\left|\begin{array}{cc}
a_1 & a_2\\
\end{array}\right|
=
\left|\begin{array}{cc}
2 & 1\\
1 & 2
\end{array}\right|
= 2^2 - 1^2 = 3
\neq 0
$$

であるため1次独立である
::::

## 線形従属（1次従属）

:::{admonition} （定義）線形従属

ベクトル$\boldsymbol{x}_1, \dots, \boldsymbol{x}_n \in V$と、すべてが0ではない係数$c_1,\dots,c_n$を用いて

$$
c_1 \boldsymbol{x}_1 + \cdots + c_n \boldsymbol{x}_n = \boldsymbol{0}
$$

とできるとき、$\boldsymbol{x}_1, \dots, \boldsymbol{x}_n$は **線形従属（linearly dependent）** あるいは **1次従属** であるという。
:::

:::{admonition} 線形従属の同値な別表現

ベクトルの組 $\boldsymbol{x}_1, \dots, \boldsymbol{x}_n$ のうち1つのベクトルがそれ以外のベクトルの線形結合で表せるとき、

例えば$\boldsymbol{x}_1$を例に取ると

$$
\boldsymbol{x}_1 = c_2 \boldsymbol{x}_2 + \cdots + c_n \boldsymbol{x}_n
$$

となるとき、$\boldsymbol{x}_1, \dots, \boldsymbol{x}_n$は線形従属である
:::

:::{dropdown} 証明

$\boldsymbol{x}_1$が他の残りのベクトルの線形結合として表示できる場合、それを

$$
\boldsymbol{x}_1 = c_2 \boldsymbol{x}_2 + \cdots + c_n \boldsymbol{x}_n
$$

とする。この式の右辺を移項すると

$$
1 \cdot \boldsymbol{x}_1 + (-c_2) \boldsymbol{x}_2 + \cdots + (-c_n) \boldsymbol{x}_n = 0
$$

となり、$\boldsymbol{x}_1$の係数は1であるため、$\boldsymbol{x}_1, \dots, \boldsymbol{x}_n$は線形従属である。


$\boldsymbol{x}_1$以外の任意のベクトルでも同様のことが言える。
:::

:::{card} （例）線形写像（行列）を使っての例

異なる入力のベクトル$\boldsymbol{x}, \boldsymbol{x}'$が同じ$\boldsymbol{y}$に移る、つまり単射でない場合、すなわち$A \boldsymbol{x} = A\boldsymbol{x}'$の場合、$A$を構成する列ベクトルたち$\boldsymbol{a}_1, \dots, \boldsymbol{a}_m$は線形従属である
:::
