# 設計（Design）

クラス〜モジュールレベルで「変更しやすいコード」をどう組み立てるかを扱う。アプリケーション全体〜システム全体の
構造は [ソフトウェアアーキテクチャ](../software_architecture/introduction.ipynb) で扱う。

## スケール別の見取り図

「良い設計とは何か」という問いへの答えは、突き詰めると[結合度を下げ凝集度を上げること](principles/coupling_and_cohesion.ipynb)
に集約される。原則・パターン・アーキテクチャは、この同じ目的を **異なるスケール** に適用したものと捉えると、
どれをどこで使うべきかが整理しやすい。

| スケール | 扱うもの | このディレクトリ内の主な内容 |
| --- | --- | --- |
| 式・関数 | 1つの関数やメソッドの書き方 | [基本原則](principles/basic_principles.ipynb)（KISS, YAGNI, DRY, SLAPなど） |
| クラス | クラス間の責務分担・依存の向き | [SOLID原則](principles/solid.ipynb)、[デザインパターン](design_patterns/index.ipynb) |
| モジュール・ドメイン | 業務知識をどうモデルへ落とすか | [ドメイン駆動設計（DDD）](domain_driven_design/index.ipynb) |
| アプリケーション | 層の分け方・依存の方向 | [ソフトウェアアーキテクチャ（application）](../software_architecture/application/index.ipynb) |
| システム・分散 | サービス間の連携・整合性 | [ソフトウェアアーキテクチャ（distributed）](../software_architecture/distributed/index.ipynb) |

コードの臭い（設計原則違反のサイン）に気づいてから改善する具体的な手順は [リファクタリング](refactoring.ipynb) にまとめている。
