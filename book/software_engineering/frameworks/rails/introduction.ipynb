### RESTful

resourcesベースにRESTfulにルーティングする

そのためResource, Controller, Modelが1:1に対応する。  
ModelとDBのテーブルはActiveRecordが1:1対応させる。

```ruby
# config/routes.rb
Rails.application.routes.draw do
  resources :posts
end
```

```ruby
# app/models/post.rb
class Post < ApplicationRecord
  validates :title, presence: true
end
```

```ruby
# app/controllers/posts_controller.rb
class PostsController < ApplicationController
  def index
    @posts = Post.order(created_at: :desc)
  end
  ...
end
```


### 密結合

- RailsのModelはユースケースと密結合
    - Modelに [validation](https://railsguides.jp/active_record_validations.html) や [callback](https://railsguides.jp/active_record_callbacks.html) などの機能を持たせている
- EntityとRepositoryの役割を担っている

→ Railsの限界：あるModelが複数の異なるユースケースでCRUD操作されるようになったとき、Railsと合わなくなる
