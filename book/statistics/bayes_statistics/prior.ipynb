### 5 levels of priors

1. Flat prior（フラット事前分布）
    - 通常は推奨されない。情報がほとんどない。
2. Super-vague but proper prior
    - 例：normal(0, 1e6)—幅が広すぎて事実上不正確。
3. Weakly informative prior（かなり弱い情報あり）
    - 例：normal(0, 10)—幅はあるがまだ情報が弱め。
4. Generic weakly informative prior（標準的な弱情報）
    - 例：normal(0, 1)—よく使われる 弱ly informative prior。
5. Specific informative prior（明確な事前情報）
    - 例：normal(0.4, 0.2)—知識や過去研究に基づく具体的な prior。

※ 各 prior の評価は パラメータのスケール に依存する（単位が適切に調整された場合に意味がある）。
