# ビジネスパーソンとしての成長について

### ビジネスパーソンの成長フェーズ

ビジネスパーソンの成長フェーズは、大別して以下のようになる

**1. 1つの仕事を期日までに十分なクオリティで納められる**

イメージとしては、新卒1~2年目のメンバーが1つの仕事をこなせるようになるまでのフェーズ。
まず仕事を1つ覚えて、きちんとできるようになるフェーズ。
更に細分化すると「言われたことができる（指示やアドバイスのもと動ける）」と「自発的に動くことができる」のフェーズも含まれる。


**2. 複数の仕事を同時に期日までに十分なクオリティで納められる**

複数のプロジェクトを推進することができるフェーズ。一定のタスク管理能力が求められる。
さらに細分化すると、

- 「自分のタスクを管理できるフェーズ（新卒2~3年目くらいのイメージ）」
- 「自分とメンバーのタスクを管理できる（チームで働くことができる）フェーズ（新卒3年目以降、チームリーダー相当）」

がある。


**3. あるべき姿（ビジョン）を描いて、他の人に浸透させて実現する**


### 職務等級

職務等級制度の例

| 階層                         | 等級 | 資格要件                                                     |
| ---------------------------- | ---- | ------------------------------------------------------------ |
| 上級職<br />（マネジメント） | 7    | 最上位資格者として、会社の経営方針に基づき、経営目標達成のための策を確定的な範囲で決定し、高い事業能力を有する。 |
|                              | 6    | 会社の経営方針に基づき、経営目標達成のための策を企画立案でき、高い事業能力を有する。 |
| 中級職<br />（シニア）       | 5    | 経営目標を理解し、上級職が立てた事業を実現するために部下をまとめ、課された目標を部下に任せながら指導・管理できる能力を有する。 |
|                              | 4    | 自ら規範を示しながら、部下をまとめ、課された目標を部下に任せながら指導・管理できる能力を有する。 |
|                              | 3    | 仕事を周りを巻き込んでチームで達成できる。                   |
| 一般職<br />（ジュニア）     | 2    | 仕事を単独で遂行する能力を有する。                           |
|                              | 1    | 指示があれば仕事をこなす能力を有する。                       |
