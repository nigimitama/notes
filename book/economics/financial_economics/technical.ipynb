# テクニカル指標

## テクニカル指標は使えるのか？

[結局のところテクニカル指標って使えるの？禁断の議論に終止符を打ちます！！ - YouTube](https://www.youtube.com/watch?v=wY2FMx5LvfY)

- 結論、使える。使い方が誤解されている
- なぜ効かないように見えるのか？ → 市況と銘柄によるスクリーニング後しか使えないから
    - ファクターをもとに運用する主体があるので、テクニカルが効く場面がある
- 例えば「過去21日騰落率」が高い銘柄/低い銘柄の差はでるが
    - マーケット全体が上がった日と下がった日で逆に傾向がでるので、マーケット全体の騰落で層別して見ないと相関が見えない
    - あるいは小型株に絞ると相関が見えてくる
