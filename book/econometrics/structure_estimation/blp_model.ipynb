# BLPモデル

単純な需要関数を推定しようとすると、財の数だけ需要関数の推定式が必要になり、財の数の2乗の数の交差弾力性を求める必要があるが、データ的にも実現可能性が低い。

[Berry et al. (1995)](http://people.stern.nyu.edu/acollard/BLP.pdf)（著者の頭文字をとってBLP）が提示した推定方法では、単純な需要関数を推定するのではなく、消費者の属性と財の特性を含む離散選択モデルにより**財の市場シェアを計算して需要を推定する**。

実際の市場シェアと予測値の市場シェアの差を最小化するようにパラメータを推定すればよいので、**集計値しかデータがなくても実施可能**な分析である。



## 需要関数の推定に関する問題（BLPの基本問題）

1. 効用関数が適切である必要
2. 財の特性と価格の間の内生性の問題
3. 分析者に観測不能な財の特性の考慮

## BLPモデルの仮定

1. 消費者の嗜好（→モデルの誤差項）はタイプ１型極値分布に従う
2. 無関係な選択対象からの独立性（IIA）：任意の２つの財の交差弾力性がそれらの財の市場シェアのみから決定され、他の財の影響を全く受けない



## ロジット型モデル

個々人の効用の違いがランダムな誤差項にのみ集約されているモデル

消費者$i$が財$j$を価格$p_j$で購入することによる効用は次のように定式化される。

$$
u_{ij} = 
\beta^1 x^1_j + \cdots + \beta^K x^K_j
- \alpha p_j + \xi_j + \varepsilon_{ij}
$$

ここで

* $x_j^1, \cdots, x_j^K$は観察可能な財の特性
* $\xi_j$は観察不可能な財の特性
* $\varepsilon_{ij}$は消費者$i$の財$j$に対する撹乱項
* $\beta^k$は財の特性に対する消費者共通の嗜好

消費者は$u_{ij} \geq u_{ij'}, \ \forall \ j' \neq j$なら財$j$を購入する。撹乱項$\varepsilon_{ij}$がそれぞれ独立にタイプ1型極値分布$P$に従うと仮定すれば、財$j$の市場シェア$s_j$は撹乱項で積分して次のように計算できる。

$$
\begin{align}
s_j
&= \int_{\{\varepsilon_{ij}: \ u_{ij} \geq u_{ij'}\}}
d P(\varepsilon_{ij})\\
&= \frac
{\exp(\beta^1 x^1_j + \cdots + \beta^K x^K_j - \alpha p_j + \xi_j)}
{1 + \sum^J_{j'=1} \exp(\beta^1 x^1_{j'} + \cdots + \beta^K x^K_{j'} - \alpha p_{j'} + \xi_{j'})}
\end{align}
$$

購入しない消費者の割合は

$$
s_0 = \frac
{1} {1 + \sum^J_{j'=1} \exp(\beta^1 x^1_{j'} + \cdots + \beta^K x^K_{j'} - \alpha p_{j'} + \xi_{j'})}
$$

となる。



## ランダム係数モデル

財の特性に対する嗜好$\beta^k$を個々人で異なるものにしたもの

$$
u_{ij} = 
\beta^1_i x^1_j + \cdots + \beta^K_i x^K_j
- \alpha p_j + \xi_j + \varepsilon_{ij}
$$

計算の容易性のため、パラメータ$\beta^1_i, \cdots, \beta_i^K$に関して以下の仮定をおく

> $\beta_i^k, i=1,...,I$は互いに独立な確率変数で、それぞれ平均$\bar{\beta}^k$、標準偏差$\sigma^k$の正規分布に従う（$k=1,...,K$）

この仮定により、集計データから$\bar{\beta}^k$や$\sigma^k$を推定することが分析の目的になる。

$\beta$は標準正規分布に従う$v$を使って消費者平均と消費者特有の部分にわけられる

$$
\beta_i^k := \bar{\beta}^k + \sigma^k v_i^k
\\
v_i^{k} \sim N(0, 1)
$$

間接効用関数も

$$
\begin{align}
u_{ij} 
&= \sum_{k=1}^K \beta^k_i x^k_j - \alpha p_j + \xi_j + \varepsilon_{ij}\\
&= \sum_{k=1}^K \hat{\beta}^k x^k_j 
- \alpha p_j + \xi_j 
+ \sum_{k=1}^K \sigma^k x_j^k v_i^k
+ \varepsilon_{ij}\\
&= \delta_j + \mu_{ij} + \varepsilon_{ij}
\end{align}
$$

となる。

市場シェアの積分は解析的に解けないので無作為抽出による数値計算で解く。

市場シェアは

$$
\hat{s}_j(\boldsymbol{\delta}, \hat{\boldsymbol{\mu}})
= \frac{1}{N} \sum^N_{n=1}
\frac
{\exp(\delta_j + \hat{\mu}_{nj})}
{1 + \sum^J_{j'=1} \exp(\delta_{j'} + \hat{\mu}_{nj'})}
\tag{4.12}
$$

となり、$\hat{\mu}_{nj}=\sum^K_{k=1}\sigma^k x^k_j v_n^k$の$v_n^k$は確率分布$F$から抽出された$N$個のサンプルである。


## BLPアルゴリズム

推定すべきパラメータを$(\boldsymbol{\bar{\beta}},\alpha)$と$\boldsymbol{\sigma}$の2つに分ける。

ステップ1：固定した$\sigma$の下での市場シェアの予測値を計算する

ステップ2：$\sigma$を固定したままで、市場シェアの予測値と実測値を等しくするような$\delta$を勾配降下法で推定する

$$
\boldsymbol{\delta}^{(\kappa+1)}
=\boldsymbol{\delta}^{(\kappa)}
+ \ln \boldsymbol{S}
- \ln \boldsymbol{s}(\boldsymbol{\delta}^{(\kappa)}, \hat{\boldsymbol{\mu}}),
\ \kappa = 0, 1, 2, ...
\tag{4.17}
$$

ステップ3：推定のための目的関数$Q$を定式化し、パラメータ$(\boldsymbol{\bar{\beta}},\alpha)$をGMMで推定する。

$$
Q(\boldsymbol{\xi}(\boldsymbol{\bar{\beta}}, \alpha))
:= \boldsymbol{m}(\boldsymbol{\xi}(\boldsymbol{\bar{\beta}}, \alpha))'
\boldsymbol{W}
\boldsymbol{m}(\boldsymbol{\xi}(\boldsymbol{\bar{\beta}}, \alpha))
$$

ここで$\boldsymbol{m}(\boldsymbol{\xi}(\boldsymbol{\bar{\beta}}, \alpha))$はモーメント、$W$はウェイト行列である。

ステップ4：新しい目的関数を定式化し、パラメータ$\boldsymbol{\sigma}$を勾配降下法で推定する。





### 制約付き最適化アルゴリズム（MPEC）

BLPは収束するまでの計算量が多い。

Dube et al. (2012)は予測値と実測値が一致すること$\hat{s}(\delta, \hat{\mu}) = S$を制約条件とした最適化として解くことを提案

### 操作変数法による内生性への対処

価格は財の特性と相関するはず。操作変数を考えなければならない。

#### 一般化モーメント法

定数項、外生変数、操作変数のベクトル$\boldsymbol{h}$と誤差ベクトル$\boldsymbol{\xi}$で標本モーメント

$$
m_\ell (\boldsymbol{\xi}(\boldsymbol{\theta}))
:= \frac{1}{TJ} \sum^T_{t=1}\sum^J_{j=1}
\xi_{tj}(\boldsymbol{\theta})\cdot h^\ell_{tj},
 \ \ell = 1,..., L
$$

を作る。この標本モーメントとは直交条件$E[\xi z] = 0$の左辺を標本平均で置き換えたものである。

#### 操作変数の選択

BLPは寡占市場のゲームに関して、ある財$j$の操作変数について、以下のような操作変数を提案している：

- 財$j$自身の特性
- 同じ企業が生産する他の財の特性
- 他の企業の財の特性
