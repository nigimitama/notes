### 疑問文

| 種類                                     | 疑問語・形式 | ピンイン      | 用法                                         | 例文                 | 日本語訳                               |
| ---------------------------------------- | ------------ | ------------- | -------------------------------------------- | -------------------- | -------------------------------------- |
| **疑問助詞文（〜吗？で文末に置く）**     | 吗           | ma            | 肯定文の語尾に付けるだけでYes/No疑問文にする | 你喜欢咖啡吗？       | あなたはコーヒーが好きですか？         |
| **反復疑問文（A不A構文）**               | A不A         | -             | 動詞や形容詞を繰り返して「〜ですか？」       | 他是不是学生？       | 彼は学生ですか？                       |
|                                          |              |               |                                              | 他是中国人，对不对？ | 彼は中国人ですよね？（そうですよね？） |
| **疑問詞文（疑問詞を使って情報を聞く）** | 什么         | shénme        | 「何？」                                     | 你在吃什么？         | あなたは何を食べていますか？           |
|                                          | 谁           | shéi / shuí   | 「誰？」                                     | 谁是你的老师？       | 誰があなたの先生ですか？               |
|                                          | 哪里 / 哪儿  | nǎlǐ / nǎr    | 「どこ？」                                   | 厕所在哪里？         | トイレはどこですか？                   |
|                                          | 什么时候     | shénme shíhou | 「いつ？」                                   | 你什么时候来？       | あなたはいつ来ますか？                 |
|                                          | 怎么         | zěnme         | 「どうやって？なぜ？」                       | 你怎么来学校？       | あなたはどうやって学校に来ますか？     |
|                                          | 为什么       | wèishénme     | 「なぜ？」                                   | 你为什么学习中文？   | なぜ中国語を勉強していますか？         |
|                                          | 哪个         | nǎge          | 「どれ？」                                   | 你喜欢哪个？         | あなたはどれが好きですか？             |
|                                          | 多大         | duō dà        | 「いくつ？何歳？」                           | 你多大了？           | あなたは何歳ですか？                   |
|                                          | 多少         | duōshao       | 「いくつ？どのくらい？」                     | 多少钱？             | いくらですか？                         |
|                                          | 怎么样       | zěnmeyàng     | 「どうですか？」                             | 这个菜怎么样？       | この料理はどうですか？                 |

### 所有代名詞

所有はN人称の代名詞と **的（de）** で表す。

| 種類          | 中国語 | ピンイン     | 日本語訳   | 例文         | 訳               |
| ----------- | --- | -------- | ------ | ---------- | --------------- |
| 一人称単数       | 我的  | wǒ de    | 私の     | 这是我的书。     | これは私の本です。       |
| 二人称単数       | 你的  | nǐ de    | あなたの   | 那是你的手机吗？   | あれはあなたの携帯ですか？   |
| 三人称単数 男性    | 他的  | tā de    | 彼の     | 这是他的电脑。    | これは彼のパソコンです。    |
| 三人称単数 女性    | 她的  | tā de    | 彼女の    | 她的名字很漂亮。   | 彼女の名前はきれいです。    |
| 三人称単数 中性    | 它的  | tā de    | それの    | 狗的尾巴是它的特点。 | 犬の尻尾はそれの特徴です。   |
| 一人称複数       | 我们的 | wǒmen de | 私たちの   | 我们的老师很严格。  | 私たちの先生は厳しいです。   |
| 二人称複数       | 你们的 | nǐmen de | あなたたちの | 你们的中文很好。   | あなたたちの中国語は上手です。 |
| 三人称複数 男性/混合 | 他们的 | tāmen de | 彼らの    | 他们的家很大。    | 彼らの家は大きいです。     |
| 三人称複数 女性のみ  | 她们的 | tāmen de | 彼女たちの  | 她们的衣服很漂亮。  | 彼女たちの服はきれいです。   |
| 三人称複数 中性    | 它们的 | tāmen de | それらの   | 它们的颜色很美。   | それらの色は美しいです。    |
