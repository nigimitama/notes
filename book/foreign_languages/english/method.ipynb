### どんな外国語でも半年でマスターしてしまう方法

[どんな外国語でも半年でマスターしてしまう方法 | クリス・ロンズデール | TEDxLingnanUniversity - YouTube](https://www.youtube.com/watch?v=d0yGdNEWdn0)

**5 Principles**

1. Focus on the language content is relevant to you
    - 自分に関係のある内容に集中せよ
2. Use your new language as a toot to communicate right from day 1
    - 初日から言葉を道具として使え。道具は使うから覚える。
3. When you first understand the message, you will unconsciously acquire the language
    - メッセージを理解すれば、自然とその言葉を習得している。
4. Physiological Training
    - 語学習得は筋トレだ！顔が筋肉痛になるくらい言葉を話す練習をする。
5. Psycho- Physiological state matters!
    - 心理状態が大事！楽しくリラックスして学習しよう。

**7 Actions**

1. Listen a lot.  Brain soaking.
    - たくさん聞け。脳を（言葉に）浸せ。
    - 意味がわからなくてもOK。聞いているうちに自然とリズムとかが身につく。
2. Focus on getting the meaning 1st (before the words)
    - 言葉より先に意味を知ること
    - ボディーランゲージとかからも意味を推測しろ
3. Start mixing
    - 学んだ単語を組み合わせて実際に表現してみる
    - 動詞10コ X 名詞 10コ X 形容詞10コで1000通りのフレーズがつくれる。
    - 完璧さより「使ってみること」を重視。
4. Focus on the core
    - コアの部分に集中すること
    - 英語だったら、  
      1000語：日常会話の85％をカバー  
      3000語：日常会話の98％をカバー  
      =3000語で英語が話せるということ
5. Get a language parent
    -  外国語の親（丁寧に言葉を教えてくれる人）を見つけろ
6. Copy the face
    - ネイティブの口や顔の動きを真似る。発音や話し方を身体で覚える。
7. Direct connect to mental images
    - 単語の意味を抽象的ではなく具体的なイメージで理解する。

**まとめ**

1. 学ぶ内容を「自分の目的」に合わせる
    - 仕事、趣味、生活場面など、実際に使う内容から始める。
2. インプットし、使う
    - 友人、交流会、言語交換パートナー、オンラインレッスンなど。
3. 学習の心的状態を整える
    - 楽しく、リラックスした状態を保ちながら進める。
