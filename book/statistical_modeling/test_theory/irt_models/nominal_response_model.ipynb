# 名義反応モデル

順序のないカテゴリカル変数を名義尺度（nominal scale）という。

名義尺度の反応を扱えるのが **名義反応モデル（nominal response model: NRM）** 。softmax関数のような形で多値化する。

$$
P(y_{ij} = k)
= \frac{ \exp (a_{jk} \theta_i + \gamma_{jk}) }{ \sum_{\kappa=1}^K \exp (- a_{j\kappa} \theta_i + \gamma_{j\kappa}) }
$$

**名義カテゴリモデル（nominal categories model）** という呼ばれ方もする
