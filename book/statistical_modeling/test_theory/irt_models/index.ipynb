# 項目反応理論のモデル群

## 概要

### 二値の順序尺度

- 正規累積モデル
- ロジスティックモデル（1PL=Rasch, 2PL, 3PLなど）

基本的にロジスティックモデル（2PL, 3PL）が使われる

:::{admonition} 2 Parameters Logistic Model

$$
P(y_{ij} = 1)
= \frac{1}{1+ \exp(-D a_j ( \theta_i - b_j))}
$$

- $a_j$：項目識別力
- $b_j$：項目困難度

:::


### 多値の順序尺度

#### 累積型

特性母数（能力母数）$\theta$が高くなるほど反応率（正答率）が高くなる、単調増加な項目反応関数を仮定するモデル

- 段階反応モデル（GRM）
- 一般化部分採点モデル（GPCM）

:::{admonition} 一般化部分得点モデル（Generalized Partial Credit Model: GPCM）

被験者 $i$、項目 $j$、カテゴリ $k = 1,\dots,K$ に対して

$$
P(y_{ij}=k)
=
\frac{
\exp\!\left(\sum_{c=1}^{k} a_j(\theta_i - b_{jc})\right)
}{
\sum_{l=0}^{K}
\exp\!\left(\sum_{c=1}^{l} a_j(\theta_i - b_{jc})\right)
}
$$

- $a_j$：項目 $j$ の識別力
- $b_{jk}$：段階 $k-1 \to k$ の困難度

:::


#### 展開型

特性母数と反応率に単調な関係がないIRTモデルを展開型（unfolding）という。例えば政治的態度や官能評価など、「好き~嫌い」の5段階評価で、回答者の好みと対象の位置の距離に比例して反応率が変わるタイプで使われる。

- 一般化段階展開型モデル(Generalized Graded Unfolding Model: GGUM)

:::
