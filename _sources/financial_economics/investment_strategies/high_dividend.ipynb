# 高配当株ファンド投資

## SCHD

```python
import yfinance as yf
import matplotlib.pyplot as plt

# SCHDのデータをダウンロード
data = yf.download('SCHD', start='2020-01-01', end='2024-12-31')

# 終値のグラフを描く
plt.figure(figsize=(10, 6))
plt.plot(data['Close'], label='SCHD Close Price')
plt.title('SCHD Closing Price from 2020 to 2024')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

## 曜日の偏りはあるか？→月曜がやや低め

```python
import yfinance as yf
import matplotlib.pyplot as plt

# SCHDのデータをダウンロード
data = yf.download('SCHD', start='2000-01-01', end='2024-12-31')
data["Date"] = data.index
data["DayName"] = data["Date"].dt.day_name()
data.groupby("DayName")["Close"].agg(["mean","median","min","max","count"])
```
