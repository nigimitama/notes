# 成功循環モデル

MITのダニエル・キム（Daniel H. Kim）が提唱した **成功循環モデル（Core Theory of Success）** は、

> 組織やチームの成果は「関係の質 → 思考の質 → 行動の質 → 結果の質」という循環構造によって高まる

と説明するマネジメント理論

## 成功循環モデルの4つの要素

このモデルは、組織の状態を以下の4つの「質」で捉える

1️⃣ 関係の質（Quality of Relationship）：信頼・心理的安全性・相互尊重の状態
- 本音を言える
- 失敗を責められない
- 立場や役割を超えて対話できる

2️⃣ 思考の質（Quality of Thinking）：問題認識・判断・意思決定の質
- 多様な意見や視点を統合できる
- 前提や仮説を疑い、構造的に考えられる
- 短期・部分最適ではなく全体・長期で考えられる

3️⃣ 行動の質（Quality of Action）：実行力と協働の質
- 自律的に動ける（指示待ちにならない）
- チーム内でスムーズに連携できる
- 小さく試し、素早く改善できる

4️⃣ 結果の質（Quality of Results）：成果・学習・価値創出の質
- 期待される成果や数値を達成できる
- 成功・失敗から学習が蓄積される
- チームや組織への信頼・評価が高まる


## 「グッドサイクル」と「バッドサイクル」

この4つの要素がどの順番で回るかによって、組織の運命は真逆になるとされる。

### 1️⃣ グッドサイクル（持続可能な成功）

「関係の質」からスタートする

- 関係の質：メンバー同士の信頼が高まると……
- 思考の質：アイデアが活発になり、自分事として考えるようになり……
- 行動の質：自発的な行動や新しい挑戦が増え……
- 結果の質：その結果として成果が出る。

→さらに「関係」が深まる（「次もこのチームで頑張ろう！」という好循環）

### 2️⃣ バッドサイクル（短期的な成果と疲弊）

「結果の質」をスタートにしてしまうと…

- 結果の質：成果が上がらないことに焦り、数字だけで詰めると……
- 関係の質：対立や責任転嫁が起き、チームの空気がギスギスし……
- 思考の質：叱られないための「言い訳」や「守りの思考」に終始し……
- 行動の質：言われたことしかやらない、消極的な行動になり……
- 結果の質：さらに結果が出なくなる。

→さらに「関係」が悪化する（負のスパイラル）

## 注意点

:::{warning}

成功循環モデルの主張は「関係の質からスタートせよ」というものだが、それへの批判として「関係が良くても目標が低い、ぬるま湯のようなゆるい組織になりかねない」という指摘がある。

目標の低い組織になることを防ぐために、「関係の質＋目標の質」でスタートするのが良いかもしれない

:::
