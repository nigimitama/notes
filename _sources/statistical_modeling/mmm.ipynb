# MMM (marketing mix modeling)

マーケティング・ミックス・モデリング（Marketing Mix Modeling: MMM）は、  売上やCV（コンバージョン）などのビジネス成果が、各マーケティング施策によってどの程度説明できるかを統計的に推定する手法。


### MMMで扱う主な変数

- **目的変数（KPI）**  
  - 売上、取扱高、来店数、サイトCV 等  
- **説明変数（施策）**  
  - TV広告（GRP）、デジタル広告（Impressions, Spend）、屋外広告、プロモーション など  
- **統制変数（外生要因）**  
  - 価格、季節性、競合指数、天候、トレンド、在庫 など


### モデル：本質的には線形回帰モデル

最も基本的な形は以下：

$$
y_t = \beta_0 + \sum_{i=1}^k \beta_i x_{i,t} + \epsilon_t
$$

しかしシンプルな線形モデルでは広告の「飽和」や「劣化（Adstock）」を表現できないため、  
**非線形変換**を施すのが一般的。


## 広告効果の非線形変換

### Adstock（広告効果の持続）

広告の効果は放送・配信した瞬間に100%消えるわけではなく、  
**徐々に減衰しながら残る**と仮定する。

代表的なモデル：  
- Geometric Adstock  
- Weibull Adstock（ベイジアンMMMでよく使われる）

例：  
$$
\text{Adstock}_t = x_t + \lambda \cdot \text{Adstock}_{t-1}
$$

### Saturation（飽和）

広告費を増やし続けても、**一定以上は効果が頭打ちになる**。

代表的関数：
- S字型（Hill関数）
- 対数関数、指数関数

例：  
$$
\text{Sat}(z) = \frac{z^\alpha}{z^\alpha + \gamma^\alpha}
$$

### モデル全体の例

$$
y_t = \beta_0 + \sum_i \beta_i \cdot \text{Sat}(\text{Adstock}(x_{i,t})) + \epsilon_t
$$


:::{admonition} [Jin, et al. (2017)](https://research.google/pubs/bayesian-methods-for-media-mix-modeling-with-carryover-and-shape-effects/) のモデル

$$
y_t=\tau+\sum_{m=1}^M \beta_m \operatorname{Hill}(x_{t, m}^* ; \mathcal{K}_m, \mathcal{S}_m)+\sum_{c=1}^C \gamma_c z_{t, c}+\epsilon_t,
$$

- $\tau$：baseline sales
- $\operatorname{Hill}(\cdot)$：広告効果の飽和関数
- $x_{t, m}^*=\operatorname{adstock}\left(x_{t-L+1, m}, \ldots, x_{t, m} ; w_m, L\right)$：広告の持続効果を考慮した出稿量
- $\gamma_c$：統制変数の回帰係数

:::

## MMMでできること

### ROI（Return on Investment）の推定
$$
\text{ROI}_i = \frac{\text{広告による増分売上}}{\text{広告費}}
$$

### ROAS（Return on Ad Spend）の推定
$$
\text{ROAS}_i = \frac{\text{増分売上}}{\text{広告費}}
$$

### MTA（Marginal Treatment Effect）の推定
最後の1円の広告費が何円の売上を生むか。

### シミュレーション & 最適化
- 予算配分最適化
- 施策A/Bの増減による売上予測

## ⚠️ MMMの課題・注意点

### 線形回帰の課題はそのままMMMの課題

- 多重共線性：複数媒体で同時に広告出稿を増やしていたら正しく推定できない
- Selection on observable：すべての交絡因子はモデルに組み込まれていると仮定する

### 広告データの難しさ

- サンプル数が非常に少ない（月次データしかない、など）
- サンプルが持つ情報（ばらつき）が少ない（例：年間予算が決まっており、毎月の広告支出が一定）
- 広告出稿に相関がある（年末商戦に向けてすべての広告媒体に予算を投下するなど）ことは多い→多重共線性
