# Pronunciation

## 母音（Vowels）

### 単母音（Monophthongs）

| 発音記号 | 例         | 備考                 |
| ---- | --------- | --------------------------- |
| æ    | cat       | 「ア」と「エ」の中間。口を大きく縦に開く。「エ」の口で「ア」の音を出すイメージ    |
| ɑ   | shot /ʃɑt/, occupation /ˌɑkjʊˈpeɪʃən/, father /ˈfɑːðə(ɹ)/ | 口を大きく開き、喉の奥から「ア」のイメージ |
| ə    | about /əˈbaʊt/, control/kənˈtɹoʊl/ | 曖昧母音。強く読まない、力を抜いて読む |
| ʌ    | up /ʌp/, come /kʌm/ | 米語だとə（シュワ）とほぼ同様だがアクセントがある。喉を開いて曖昧に |
| ɜ    | first /fɜːrst/, work /wɜːrk/ | 米語だとrが続く音のとき出る |
| ɪ    | hill /hɪl/, sit /sɪt/ | 「イ」と「エ」の中間。口をリラックスして開いた状態で「イ」を出す  |
| iː   | see /ˈsiː/, sheep /ʃiːp/ | 日本語の「イ」に近い。「イ」より口を強く横に引き、長く伸ばす |
| ɒ    | hot (BrE) | 日本語の「オ」より口を丸く低く         |
| ʊ    | hood /hʊd/, book /bʊk/, put /pʊt/ | 「ウ」と「オ」の中間。喉の奥から「オ」に近づけつつ「ウ」を言うイメージ |
| uː   | food /fuːd/      | 日本語の「ウ」に近いが唇を強くすぼめて長く     |
| e/ɛ    | bed /bɛd/       | 日本語の「エ」に近いが口をやや横に       |
| ɔː   | all /ɔːl/, law /lɔ/ | 「オー」だが**唇をしっかり丸める**         |


### 二重母音（Diphthongs）

| 発音記号 | 例   | 備考          |
| ---- | --- | -------------------- |
| aɪ   | my /maɪ/  | 「アイ」だが一音節で滑らかに   |
| aʊ   | now /naʊ/ | 「ア→ウ」。途中で口をすぼめる  |
| eɪ   | say /seɪ/ | 長母音「エー」ではなく「エィ」 |
| ɔɪ   | boy /bɔɪ/ | 「オ→イ」。 |
| oʊ   | go /ɡoʊ/ | 長母音「オー」ではなく「オゥ」。例えばboat/bəʊt/を「オー」にするとbought/bɔːt/のように聞こえる |


## 子音（Consonants）

子音は空気をせき止めたりする音

:::{card}

有声音と無声音がある

- **有声音**：喉（声帯）が震える
- **無声音**：喉（声帯）が震えない

有声音 / 無声音のペア（例えば /p/ と /b/）のセットで覚えると楽

:::



### 破裂音

#### /p/ /b/

| 発音記号 | 例    | 備考        |
| ---- | ---- | ------------------ |
| p    | pen, play  | （無声音）息を強く出す。破裂音    |
| b    | bad, book  | （有声音）息は弱め           |

#### /t/ /d/

| 発音記号 | 例    | 備考        |
| ---- | ---- | ------------------ |
| t    | top /tάp/  | （無声音）日本語「タ」より**舌を歯茎に** |
| d    | dog /dˈɔːg/ | （有声音）日本語「ダ」より**舌の位置が前** |

#### /k/ /g/

| 発音記号 | 例    | 備考        |
| ---- | ---- | ------------------ |
| k    | cat /kˈæt/ | （無声音）日本語「カ」より**息を強く**   |
| g    | dog /dˈɔːg/ | （有声音）日本語「ガ」より**濁りを明確に** |


### 摩擦音

#### /f/ /v/

| 発音記号 | 例    | 備考        |
| ---- | ---- | ------------------ |
| f    | fly /flaɪ/ | 下唇の内側を前歯の先に軽く当てた隙間から出る音（無声音） |
| v    | view /vjuː/ | 下唇の内側を前歯の先に軽く当てて、強めに発音し喉を震わせる（有声音）|

#### /θ/ /ð/

| 発音記号 | 例    | 備考        |
| ---- | ------- | ----------------- |
| θ    | think   | 舌を歯に軽く挟んで隙間から出す音（無声音） |
| ð    | this    | θの音を濁らせる（有声音） |


#### /s/ /z/

| 発音記号 | 例    | 備考        |
| ---- | ------- | ----------------- |
| s    | see     | 日本語より**鋭く強い**     |
| z    | zoo     | 「ズ」ではなく**摩擦を保つ**  |

#### /ʃ/ /ʒ/

| 発音記号 | 例    | 備考        |
| ---- | ------- | ----------------- |
| ʃ    | ship /ʃɪp/ | 唇を丸める。「シーッ！静かに」という時のシの音 |
| ʒ    | vision /ˈvɪʒ.ən/ | ʃの口から喉を震わせる（有声音）      |

- sea /siː/ と she /ʃi/ の混同に注意


#### /tʃ/ /dʒ/

| 発音記号 | 例      | 備考         |
| ---- | ------ | ------------------- |
| tʃ   | church /t͡ʃɜːt͡ʃ/, charge /t͡ʃɑːd͡ʒ/  | 「チ」より勢いがある発音。 t＋ʃ      |
| dʒ   | judge /d͡ʒʌd͡ʒ/, Japan /d͡ʒəˈpæn/ | 有声音バージョン。d＋ʒ      |



### /r/ と /l/


| 発音記号 | 例      | 備考         |
| ---- | ------ | ------------------- |
| r    | red    | 舌を巻くというよりは「舌先をどこにも当てない」「舌を少し引っ込める」、lに比べると濁ったような音 |
| l    | like /laɪ̯k/, love /lʌv/ | 舌先を**歯茎にしっかり当てる**   |

:::{card} "light L" と "dark L"

- light L：舌先は上の前歯に必ず当たる
    - like /laɪ̯k/, hello /hɛˈloʊ/ など母音が続くとき
- dark L：舌先は必ずしも上の前歯に当たるわけではない。近い位置にいればいい
    - real /ɹeɪˈɑl/, people /ˈpipəl/　など

:::

## 要点まとめ

:::{admonition} 重要項目

母音：ɪ / æ / ʌ / ə を区別できるか  
子音：r / l / θ / ð / v / f を日本語化していないか  
弱形（ə）を強く読まない

:::
