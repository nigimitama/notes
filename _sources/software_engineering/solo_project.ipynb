# 個人開発

:::{card} [個人開発の生産性を上げる5つの視点と32の工夫 #エンジニア - Qiita](https://qiita.com/Akira-Isegawa/items/ab7d6b5db65dc830bb62)

- 週に20時間以上を確保する（例：平日は2時間、土曜に10時間）
- まずは一番何をやりたいかを明確にして、そこにやる気とエネルギーを集中させる
- 1週間に20個以上のアイディアを出し、それを書き留める
- プロトタイプを素早く作って、実際に使ってもらう

:::
