# スタック

**スタック（stack）** はデータの中で最後に入ったものが最初に取り出される **後入れ先出し（Last In First Out: LIFO）** のデータ構造

操作：

- `push(x)`: スタックに要素`x`を追加する
- `pop()`: スタックの最後の要素を取り出す

```python
class Stack:

    def __init__(self):
        self.data = []

    def push(self, x):
        self.data.append(x)

    def pop(self):
        return self.data.pop(-1)


s = Stack()
s.push(1)
s.push(2)
s.pop()
```
