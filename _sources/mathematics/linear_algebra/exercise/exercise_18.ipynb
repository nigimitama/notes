# 練習問題メモ 18（表現行列）

## 問1

:::{card}

$$
\begin{aligned}
& \boldsymbol{a}_1, \boldsymbol{a}_2, \boldsymbol{a}_3 \in \mathbb{R}^3 、 \boldsymbol{b}_1, \boldsymbol{b}_2 \in \mathbb{R}^2 \text { を } \\
& \boldsymbol{a}_1=\left(\begin{array}{l}
0 \\
1 \\
2
\end{array}\right), \boldsymbol{a}_2=\left(\begin{array}{l}
1 \\
0 \\
3
\end{array}\right), \boldsymbol{a}_3=\left(\begin{array}{l}
2 \\
3 \\
0
\end{array}\right), \boldsymbol{b}_1=\binom{1}{2}, \boldsymbol{b}_2=\binom{3}{4}
\end{aligned}
$$

により定めると、 $\left\{\boldsymbol{a}_1, \boldsymbol{a}_2, \boldsymbol{a}_3\right\},\left\{\boldsymbol{b}_1, \boldsymbol{b}_2\right\}$ はそれぞれ $\mathbb{R}^3, \mathbb{R}^2$ の基底である。線形写像 $f: \mathbb{R}^3 \rightarrow \mathbb{R}^2$ 
を

$$
f(\boldsymbol{x})=\left(\begin{array}{lll}
1 & 0 & 3 \\
0 & 2 & 0
\end{array}\right) \boldsymbol{x} \quad\left(\boldsymbol{x} \in \mathbb{R}^3\right)
$$

により定める。基底 $\left\{\boldsymbol{a}_1, \boldsymbol{a}_2, \boldsymbol{a}_3\right\},\left\{\boldsymbol{b}_1, \boldsymbol{b}_2\right\}$ に関する $f$ の表現行列を求めよ。
:::

表現行列は

$$
\left(f\left(\boldsymbol{a}_1\right), \cdots, f\left(\boldsymbol{a}_n\right)\right)
=\left(\boldsymbol{b}_1, \cdots, \boldsymbol{b}_m\right) P
$$

のような行列$P$のこと

今回は、

$$
\begin{pmatrix}f(\boldsymbol{a}_1) & f(\boldsymbol{a}_2) & f(\boldsymbol{a}_3)\end{pmatrix}
= \begin{pmatrix}\boldsymbol{b}_1 & \boldsymbol{b}_2\end{pmatrix} P
$$

とおき、表現行列$P$求める。

$\begin{pmatrix}f(\boldsymbol{a}_1) & f(\boldsymbol{a}_2) & f(\boldsymbol{a}_3)\end{pmatrix}$については、

$$
f(\boldsymbol{a}_1) = 
\left(\begin{array}{lll}
1 & 0 & 3 \\
0 & 2 & 0
\end{array}\right)
\begin{pmatrix}0 \\ 1\\ 2\end{pmatrix}
= \begin{pmatrix}6 \\ 2\end{pmatrix}
\\
f(\boldsymbol{a}_2) = \begin{pmatrix}10 \\ 0\end{pmatrix}
\\
f(\boldsymbol{a}_3) = \begin{pmatrix}2 \\ 6\end{pmatrix}
$$

なので

$$
\begin{pmatrix}f(\boldsymbol{a}_1) & f(\boldsymbol{a}_2) & f(\boldsymbol{a}_3)\end{pmatrix}
= \begin{pmatrix}\boldsymbol{b}_1 & \boldsymbol{b}_2\end{pmatrix} P
\\
\iff
\begin{pmatrix}
6& 10 & 2\\
2 & 0 & 6
\end{pmatrix}
= \begin{pmatrix}
1 & 3\\
2 & 4
\end{pmatrix} P
$$

を解いて$P$を求めることになる

$$
B:=
\begin{pmatrix}
1 & 3\\
2 & 4
\end{pmatrix}
$$

の逆行列を求める。拡大係数行列を

$$
(B | E) =
\left(\begin{array}{cc|cc}
1 & 3 & 1 & 0\\
2 & 4 & 0 & 1
\end{array}\right)
$$

とおいて変形すると、

1行目を2倍して2行目から引いて

$$
\left(\begin{array}{cc|cc}
1 & 3 & 1 & 0\\
0 & -2 & -2 & 1
\end{array}\right)
$$

2行目を-1/2倍して

$$
\left(\begin{array}{cc|cc}
1 & 3 & 1 & 0\\
0 & 1 & 1 & -\frac{1}{2}
\end{array}\right)
$$

2行目を3倍して1行目から引いて

$$
\left(\begin{array}{cc|cc}
1 & 0 & -2 & \frac{3}{2}\\
0 & 1 & 1 & -\frac{1}{2}
\end{array}\right)
$$

よって

$$
B^{-1}=
\left(\begin{array}{cc}
 -2 & \frac{3}{2}\\
 1 & -\frac{1}{2}
\end{array}\right)
$$


先程の式の両辺に左から乗じると

$$
B^{-1}
\begin{pmatrix}
6& 10 & 2\\
2 & 0 & 6
\end{pmatrix}
=
B^{-1}
\begin{pmatrix}
1 & 3\\
2 & 4
\end{pmatrix} P
$$

すなわち

$$
\begin{pmatrix}
 -2 & \frac{3}{2}\\
 1 & -\frac{1}{2}
\end{pmatrix}
\begin{pmatrix}
6& 10 & 2\\
2 & 0 & 6
\end{pmatrix}
=
\begin{pmatrix}
-9 & -20 & 5\\
5 & 10 & -1
\end{pmatrix}
= P
$$


よって基底 $\left\{\boldsymbol{a}_1, \boldsymbol{a}_2, \boldsymbol{a}_3\right\},\left\{\boldsymbol{b}_1, \boldsymbol{b}_2\right\}$ に関する$f$の表現行列は

$$
\begin{pmatrix}
-9 & -20 & 5\\
5 & 10 & -1
\end{pmatrix}
$$

```python
import numpy as np

A = np.array([
    [-2, 3/2],
    [1, -1/2],
])

B = np.array([
    [6, 10, 2],
    [2, 0 , 6]
])


A @ B
```
