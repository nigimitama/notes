# 概要 - DFLとは

**Decision-Focused Learning (DFL)** は、予測性能ではなく、最終的な意思決定の質の最大化（コストを最小化）するようにモデルを訓練させる考え方。  
逆に、予測性能を上げること自体に主眼を置いた通常の機械学習は **Prediction-focused Learning (PFL)** と呼ばれる。

## 用語

### Prediction-focused Learning (PFL)



### Predict-Then-Optimize (PTO)

意思決定の問題を「まず予測し、そして最適化問題を解く」という2段階を独立して扱うアプローチのことを **Predict-Then-Optimize (PTO)** と呼ぶ。
このアプローチは「良い予測精度がよい意思決定をもたらす」という暗黙の仮定のもとで各ステップに注力するが、予測のステップで過学習してしまい局所最適解に陥るリスクがある。


### Decision-Focused Learning (DFL)

意思決定の最適化問題を、予測の損失関数とつなげた形で一気通貫して扱うアプローチを **Decision-Focused Learning (DFL)** と呼ぶ
