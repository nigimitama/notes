# Lead Scoring

## 用語

### Leadとは

Lead（リード）とは、将来的に顧客になる可能性のある個人または企業を指す。たとえば、Webサイトから資料請求をした人、問い合わせフォームを送信した企業、セミナーに参加した見込み顧客などがLeadに該当する。

Leadは、まだ購入や契約に至っていない段階の顧客候補であり、営業・マーケティング活動において重要な管理対象となる。Leadを適切に管理することで、関心度の高い顧客候補を見極め、効率的な営業活動につなげることができる。

### Lead Scoringとは

Lead Scoring（リードスコアリング）とは、各Leadに対して「どの程度顧客化する可能性が高いか」を数値化する手法。  
Leadの属性情報や行動履歴をもとにスコアを付与し、営業やマーケティングの優先順位付けに活用する。

Lead Scoringの目的は、営業リソースを効率的に配分し、成約可能性の高いLeadに適切なタイミングでアプローチすることである。これにより、営業活動の生産性向上、マーケティング施策の改善、顧客獲得率の向上が期待できる。
